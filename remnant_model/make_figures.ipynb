```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv('results/kfold_remnant_predictions.csv', dtype=str)
b = pd.read_csv('input_data/mass_doe_data.csv', dtype=str)[['school_code', 'charter_school']]
a = a.merge(b, how='inner', on='school_code')
```

```python
plt.figure()
plt.scatter(a['actual_performance'].astype(float), a['predicted_performance'].astype(float), c=a['charter_school'].astype(float))
plt.xlabel('Actual Performance')
plt.ylabel('Predicted Performance')
plt.title('Yellow = Charter School')
plt.show()
```

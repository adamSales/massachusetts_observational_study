```python
import time
import numpy as np
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
```

```python
def scrape_a(soup):
    return_dict = dict()
    super_table = soup.find('table')
    for sub_table in super_table.find_all('table')[:3]:
        title = sub_table.find('th').text
        table = pd.read_html(str(sub_table), header=1, index_col=0)[0].dropna(axis=1)
        for row_name, series in table.iterrows():
            for col_name, value in series.iteritems():
                if title == 'Enrollment by Grade (2017-18)' and row_name != 'District':
                    row_name = 'School'
                try:
                    return_dict[f'{title}__{row_name}__{col_name}'] = float(value)
                except:
                    return_dict[f'{title}__{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_b(soup):
    return_dict = dict()
    super_table = soup.find('table')
    sub_table = super_table.find('table')
    table = pd.read_html(str(sub_table), header=0, index_col=0)[0]
    for row_name, series in table.iterrows():
        for col_name, value in series.iteritems():
            try:
                return_dict[f'{row_name}__{col_name}'] = float(value)
            except:
                return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_c(soup):
    return_dict = dict()
    super_table = soup.find('table')
    sub_table = super_table.find('table')
    table = pd.read_html(str(sub_table), header=0, index_col=0)[1]
    for row_name, series in table.iterrows():
        for col_name, value in series.iteritems():
            try:
                return_dict[f'{row_name}__{col_name}'] = float(value)
            except:
                return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_d(soup):
    return_dict = dict()
    sub_table = soup.find_all('table')[2]
    table = pd.read_html(str(sub_table), header=0, index_col=0)[0]
    for row_name, series in table.iterrows():
        for col_name, value in series.iteritems():
            try:
                return_dict[f'{row_name}__{col_name}'] = float(value)
            except:
                return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_e(soup):
    return_dict = dict()
    sub_table = soup.find_all('table')[2]
    table = pd.read_html(str(sub_table), header=0, index_col=0)[0]
    for row_name, series in table.iloc[3:].iterrows():
        for col_name, value in series.iteritems():
            try:
                return_dict[f'{row_name}__{col_name}'] = float(value)
            except:
                return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_f(soup):
    return_dict = dict()
    super_table = soup.find('table')
    for sub_table in super_table.find_all('table')[:2]:
        table = pd.read_html(str(sub_table), header=0, index_col=0)[0].dropna(axis=1)
        for row_name, series in table.iterrows():
            for col_name, value in series.iteritems():
                value = value.split(' to')[0] if 'to' in str(value) else value
                try:
                    return_dict[f'{row_name}__{col_name}'] = float(value)
                except:
                    return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_g(soup):
    return_dict = dict()
    super_table = soup.find('table')
    title = super_table.find('h2').text.split(' (')[0]
    sub_table = super_table.find('table')
    table = pd.read_html(str(sub_table), header=0, index_col=0)[0]
    for row_name, series in table.iterrows():
        for col_name, value in series.iteritems():
            try:
                return_dict[f'{title}__{row_name}__{col_name}'] = float(value)
            except:
                return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_h(soup):
    return_dict = dict()
    super_table = soup.find('table')
    sub_table = super_table.find('table')
    table = pd.read_html(str(sub_table), header=[0,1], index_col=0)[0]
    for row_name, series in table.iterrows():
        for col_name, value in series.iteritems():
            try:
                col_name = ' '.join(col_name)
                return_dict[f'Staffing Retention Rates__{row_name}__{col_name}'] = float(value)
            except:
                return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_i(soup):
    return_dict = dict()
    super_table = soup.find('table')
    sub_table = super_table.find('table')
    table = pd.read_html(str(sub_table), header=0, index_col=0)[0]
    for row_name, series in table.iterrows():
        for col_name, value in series.iteritems():
            try:
                return_dict[f'Educator Evaluation Performance Rating__{row_name}__{col_name}'] = float(value)
            except:
                return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict

def scrape_j(soup):
    return_dict = dict()
    super_table = soup.find('table')
    title = super_table.find('h2').text.split('Percent')[0]
    sub_table = super_table.find_all('table')[1]
    table = pd.read_html(str(sub_table), header=[0,1], index_col=0)[0]
    for row_name, series in table.iterrows():
        for col_name, value in series.iteritems():
            try:
                col_name = ' '.join(col_name)
                return_dict[f'{title}__{row_name}__{col_name}'] = float(value)
            except:
                return_dict[f'{row_name}__{col_name}'] = float('nan')
    return return_dict
```

```python
scores6 = pd.read_excel('mcas_exports/NextGenMCAS6.xlsx', header=1, dtype=str)
scores7 = pd.read_excel('mcas_exports/NextGenMCAS7.xlsx', header=1, dtype=str)
scores8 = pd.read_excel('mcas_exports/NextGenMCAS8.xlsx', header=1, dtype=str)
scores = pd.concat([scores6, scores7, scores8], axis=0)
scores = scores[scores['School Name'] != 'State']
scores = scores[scores['Subject'] == 'MATH']
scores['Avg. Scaled Score'] = scores['Avg. Scaled Score'].astype(float)
scores['No. of Students Included'] = scores['No. of Students Included'].astype(float)
scores = scores.groupby('School Code')
scores = scores.apply(lambda x: np.average(x['Avg. Scaled Score'], weights=x['No. of Students Included']))
scores = scores.rename('Avg. Scaled Score').reset_index()
print(f'There are {len(scores)} schools in the dataset.')
print(f'The data pull is expected to take over {len(scores) * 18 * 10 / 3600} hours.')
```

```python
for school_year in [2012, 2013]:
    all_dicts = []
    for school_id, score in zip(scores['School Code'], scores['Avg. Scaled Score']):
        print(school_id)
        urls = [
            f'https://profiles.doe.mass.edu/profiles/student.aspx?orgcode={school_id}&orgtypecode=6&fycode={school_year}&leftNavId=300',
            f'https://profiles.doe.mass.edu/profiles/student.aspx?orgcode={school_id}&orgtypecode=6&fycode={school_year}&leftNavId=16817',
            f'https://profiles.doe.mass.edu/profiles/student.aspx?orgcode={school_id}&orgtypecode=6&fycode={school_year}&leftNavId=305',
            f'https://profiles.doe.mass.edu/mobility/default.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6',
            f'https://profiles.doe.mass.edu/students/classsizebygenderpopulation.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6',
            f'https://profiles.doe.mass.edu/students/classsizebyraceethnicity.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6',
            f'https://profiles.doe.mass.edu/profiles/teacher.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6',
            f'https://profiles.doe.mass.edu/profiles/teacher.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6&leftNavId=817',
            f'https://profiles.doe.mass.edu/profiles/teacher.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6&leftNavId=828',
            f'https://profiles.doe.mass.edu/profiles/teacher.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6&leftNavId=829',
            f'https://profiles.doe.mass.edu/profiles/teacher.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6&leftNavId=830',
            f'https://profiles.doe.mass.edu/profiles/teacher.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6&leftNavId=15619',
            f'https://profiles.doe.mass.edu/profiles/teacher.aspx?orgcode={school_id}&fycode={school_year}&orgtypecode=6&leftNavId=12505',
            f'https://profiles.doe.mass.edu/mcas/achievement_level.aspx?linkid=32&orgcode={school_id}&orgtypecode=6&fycode={school_year-5}',
            f'https://profiles.doe.mass.edu/mcas/achievement_level.aspx?linkid=32&orgcode={school_id}&orgtypecode=6&fycode={school_year-4}',
            f'https://profiles.doe.mass.edu/mcas/achievement_level.aspx?linkid=32&orgcode={school_id}&orgtypecode=6&fycode={school_year-3}',
            f'https://profiles.doe.mass.edu/mcas/achievement_level.aspx?linkid=32&orgcode={school_id}&orgtypecode=6&fycode={school_year-2}',
            f'https://profiles.doe.mass.edu/mcas/achievement_level.aspx?linkid=32&orgcode={school_id}&orgtypecode=6&fycode={school_year-1}',
        ]
        functions = [
            scrape_a,
            scrape_b,
            scrape_b,
            scrape_c,
            scrape_d,
            scrape_e,
            scrape_f,
            scrape_g,
            scrape_g,
            scrape_g,
            scrape_g,
            scrape_h,
            scrape_i,
            scrape_j,
            scrape_j,
            scrape_j,
            scrape_j,
            scrape_j
        ]
        combined_dict = {'school_id': school_id, 'Avg. Scaled Score': score}
        for url, func in zip(urls, functions):
            done = False
            while not done:
                try:
                    html = urllib.request.urlopen(url).read()
                    bs = BeautifulSoup(html, 'lxml')
                    combined_dict.update(func(bs))
                    done = True
                except Exception as e:
                    print(e)
                time.sleep(10)
        all_dicts.append(combined_dict.copy())
        df = pd.DataFrame.from_records(all_dicts)
        df.to_csv(f'mass_doe_data_{school_year}.csv', index=False)
```
